# Accessing Items from a List

### Indexing

```python
L = [1,2,3,4,5]
#Positive Indexing
print(L[1])
#Negative Indexing
print(L[-1])
```

```python
#2D List
L = [1,2,3,[4,5]]
print(L[-1])
```

```python
# 3D List
L = ([[[1,2],[3,4]],[[5,6],[7,8]]])
print(L[0])
print(L[0][1][0])
print(L[0][0][1])
```

### Slicing

```python
L = [1,2,3,4,5]
print(L[0:3])
print(L[-3:-1])
print(L[-3:])
print(L[0:])
print(L[0::2])
print(L[-5:-2:2])
print(L[::-1])
```

# Adding Items to a List

### append
You can add single value at last in the list

```python
L = [1,2,3,4,5]
L.append(6)
print(L)
```

### extend
you can add multiple values at last in the list

```python
L = [1,2,3,4,5]
L.extend([6,7,8])
print(L)
```

```python
L = [1,2,3,4,5]
L.extend("delhi")
print(L)
```

##### Difference between extend and append

```python
L = [1,2,3,4,5,6]
L.extend([7,8])
print(L)
L.append([7,8])
print(L)
```

### Insert
It can add the value at any index of the list

```python
L = [1,2,3,4,5]
L.insert(1,100)
print(L)
```

```python
L = [1,2,3,4,5]
L.insert(1,[100,200])
print(L)
```

# Editing items in a List

String are immutable

```python
s = 'hello'
s[0]='x'
print(s)
```

But the List are mutable

```python
L = [1,2,3,4,5]
#editing with indexing
L[-1]=500
print(L)
#editing with slicing
L[1:4]=[200,300,400]
print(L)
```

# Deleting items from a List

### del

```python
L = [1,2,3,4,5]
print(L)
del L # it delete all the list
print(L)
```

```python
L = [1,2,3,4,5]
#delete with indexing
del L[-1]
print(L)
#delete with slicing
del L[1:2]
print(L)
```

### remove


```python
L = [1,2,3,4,5]
L.remove(5) #it doesn't require to know index value but it require just the value present in the list
print(L)
```

### pop

```python
L = [1,2,3,4,5]
print(L)
L.pop(0)
print(L)
L.pop() # it delete the last item if the index is not provide
print(L)
```

### clear

```python
L = [1,2,3,4,5]
L.clear() # it clear the value inside the list
print(L)
```

# List Comprehension

List Comprehension provides a concise way of creating lists.

newlist = [expression for item in iterable if condition == True]



Advantages of List Comprehension
- More time-efficient and space-efficient than loops.
- Require fewer lines of code.
- Transforms iterative statement into a formula.

# Example

### Add 1 to 10 number to a list

```python
# Normal program
L = []
for i in range(1,11):
    L.append(i)
print(L)
```

```python
#List Comprehension
L = [i for i in range(1,11)]
```

### scalar multiplication on a vector

```python
v = [2,3,4]
s = -3
[s*i for i in v]
```

### Add squares

```python
L = [1,2,3,4,5]
[i**2 for i in L]
```

### Print all number divisible by 5 in the range of 1 to 50

```python
print([i for i in range(1,51) if i%5==0])
```

### find languages which start with letter p

```python
languages = ['java','python','php','c','javascript']
[language for language in languages if language.startswith('p')]
```

```python
languages = ['java','python','php','c','javascript']

[i for i in languages if i.startswith('p')]
```

### Nested if with List Comprehension
add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

```python
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

### Print a (3,3) matrix using list comprehension -> Nested List comprehension

```python
[[i*j for i in range(1,4)] for j in range(1,4)]
```

### cartesian products -> List comprehension on 2 lists together

```python
L1 = [1,2,3,4]
L2 = [5,6,7,8]
[i*j for i in L1 for j in L2]
```

## 2 ways to traverse a list

- itemwise
- indexwise

itemwise

```python
L = [1,2,3,4,5]
for i in L:
    print(i)
```

indexwise

```python
L = [1,2,3,4,5]
for i in range(0,len(L)):
    print(i) #index of the item
```

```python
L = [1,2,3,4,5]
for i in range(0,len(L)):
    print(L[i])
```

## Zip

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]
# zip(L1,L2)  ## it gives zip object which can change in list by using list(zip())
list(zip(L1,L2))
[i+j for i,j in zip(L1,L2)]
```
